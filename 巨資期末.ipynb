```python
import pandas as pd
import csv
import time
import matplotlib.pyplot as plt
import re
```

匯入levelsmap_concat.csv\
轉時間+清理ID\
排序

```python
test = pd.read_csv("levelsmap_concat.csv")
#轉時間
test['start_date_new']=test['startTimeInSeconds'].apply(
    lambda x: time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(x+8*3600)) if not(pd.isnull(x)) else x)
test['end_date_new']=test['endTimeInSeconds'].apply(
    lambda x: time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(x+8*3600)) if not(pd.isnull(x)) else x)
test["new_id"]=test["id"].apply(lambda x:x.replace("levelsmap_",""))
test=test.sort_values(by=['new_id','start_date_new'])

#選定一個人和時間，看時間是否對齊
# id1 = test[test['new_id']=='1091_uty_1']
# id1 = id1[id1['date']==1017.0]
# id1.sort_values(by=['start_date_new'])
```

清理時間

```python
test['date_new']=test['start_date_new'].apply(
    lambda x: re.search('\d{4}\-\d{2}\-\d{2}',x).group())
test
```

睡眠時長

```python
test['last_time']=test['endTimeInSeconds']-test['startTimeInSeconds']
#test
```

刪除重複值

```python
test=test.drop_duplicates(subset=["startTimeInSeconds"],keep="first")
test
```
